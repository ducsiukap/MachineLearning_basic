```python
import pandas as pd
# Series is 1D NumPy-like array,  nhưng có thêm thuộc tính index cho từng phần tử (mặc định là 0, 1, 2, ...)
```

```python
# create a series

# using default index
series =  pd.Series([1, 2, 3, 4, 5])
print(f'index value\n{series}')

# using specified index
series = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
print(f'\nindex value\n{series}')

# tạo series từ dictionary
d = {'a': 1, 'b': 2, 'c': 3}
series = pd.Series(d)
print(f'\nindex value\n{series}')

# specify datetime range as Index - pd.date_range(startdatetime, periods, freq='D') #yyyymmdd, yyyymmdd hhmmss, ... 
# or: pd.date_range(startdatetime, enddatetime, freq='D')
date_indexs = pd.date_range('20250822', periods=3) 
print(f'\n>>> date_indexs:\n{date_indexs}')
series.index = date_indexs
print(f'index value\n{series}')
# change the freq
print(pd.date_range('202508221706', periods=3)) # default, freq='D' - Daily
print(pd.date_range('202508221706', periods=3, freq='ME')) # Month
print(pd.date_range('202508221706', periods=3, freq='h')) # Month
# freq meaning
# B -> business day (Mon -> Fri)
# W -> Week (Ngày kết thúc của tuần, mặc định là cuối tuần - sunday) 
# W-MON, W-TUE 
# ME, MS -> Month End / Month Start
# QE, QS -> Quarter End / Quarter Start
# YE, YS -> Year end / start
# t, min -> Minute
# s -> second
# h -> hour
# ...
```

```python
# accessing elem

s1 = pd.Series([1, 2, 3, 4, 5], index=[3, 1, 2, 0, 5]) # index là số nguyên
s2 = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
s3 = pd.Series([1, 2, 3, 4, 5]) # 
print(f'index value\n{s1}')
print(f'index value\n{s2}')
print()

# access using [label] op
# nếu không định nghĩa thì index của series có dạng: index=[0, 1, 2, ...] 
print(s1[0], s1[5]) # lấy phần tử có label là 0
# print(s1[4]) # error
print(s2['a'], s2['c'], s2['d'])
# print(s2[0]) # error
print()

# label-based with .loc(label)
print(s1.loc[1], s1.loc[3])
print(s2.loc['b'])
print()

# position-based with .iloc(position) -> 0, 1, 2, ...
print(s1.iloc[0], s1.iloc[3])
print(s2.iloc[2], s2.iloc[4])

```

```python
# dù [] op sử dụng label
# nhưng
# slice: [start:end:step] lại dùng position :))

series = pd.Series([1, 2, 3, 5, 4, 6], index=[100, 200, 300, 400, 500, 600])
print(f'index value\n{series}')

print(f'\n>>> series[1:4] : ')
print(f'index value\n{series[1:4]}')

# reverse 
print('\n', series[::-1], sep='')
print()

# slicing sử dụng label dùng .loc[:]
series = pd.Series([1, 2, 3, 4, 5], ['a', 'b', 'e', 'd', 'c'])
print(series.loc['a':'e'])
```

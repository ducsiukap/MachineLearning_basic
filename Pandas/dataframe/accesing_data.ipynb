# Select specified rows using loc() and iloc()

```python
df2 = df.copy()
df2.index = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K']
df2
```

```python
# loc: using label
df2.loc[['A', 'E']] # index=[A, E]
```

```python
# select row  & column
df2.loc['A':'B', 'A':'C'] # select row A, B. each row select column A, B, C
```

```python
# iloc : using integer (natural) index 0, 1, 2... 
df2.iloc[[2, 5]] #integer index
```

```python
# specified row & col
df2.iloc[[0, 4, 9], [0, 2, 3]] # select row 0, 4, 9 (A, E, K). each row select column 0, 2, 3 (A, C, D)
```

# Set values using loc[] = value , iloc[] = value

```python
df2.loc['B':'E', 'A'] = [2, 3, 4, 5] # set row from B->E, col A by values: 2, 3, 4 ,5
df2.loc['F':'G', 'B'] = -1 # set row F->G, col B by values = [-1, -1]

df2.iloc[0] = [1, 2, 3, 4] # set row 1 by values (1, 2, 3, 4)
df2.iloc[7:10, 3] = -99 # set row 7->8, column 3 (D) by values [-99, -99, -99]

df2
```

# select one specific item with at() and iat()

```python
df2.at['A', 'B'] # row A col B 
```

```python
df2.iat[0, 1] # row A col B
```

# Adding/Removing column
` df[new_columns] = [values] `

### adding column

```python
# create new column
df['author'] = 'vduczz'

# create new column base on exist one
df['positive_b'] = np.where(df.B > 0, True, False)
df.head()
```

### removing rows

```python
# removing rows

# base on label
dropped = df.drop(3)
dropped
```

```python
# filter to drop
dropped = df[df['positive_b'] == True]
dropped
```

```python
# drop using index
dropped = df.drop(df.index[[1, 3, 4]])
dropped
```

### removing column
`drop(axis=1)`

```python
# using column name
dropped = df.drop(['C', 'D'], axis=1)
dropped
```

```python
# using column index
dropped = df.drop(df.columns[0], axis=1)
dropped
```

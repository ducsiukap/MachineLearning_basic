## **Generating Dataset** - `sklearn.datasets`

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_regression, make_classification, make_blobs, make_circles
# make_regression, make_classification, make_blobs, ... 

# Linearly Distributed Datase
X, y = make_regression(n_samples=100, n_features=3, noise=5.4)
# n_features -> số feature
# 3features, 1target
plt.subplot(1, 3, 1)
plt.scatter(X[:, 0], y, c='r', marker='D') # feature1-target
plt.scatter(X[:, 1], y, c='g', marker='x') # feature2-target
plt.scatter(X[:, 2], y, c='b', marker='^') # feature3-target
print(X[:5], y[:5], sep='\n')

# Clustered Dataset -> Dữ liệu giả lập cho bài toán phân cụm -> unsupervised learning (không label, tự tìm cụm phù hợp nhất)
X, y = make_blobs(n_samples=100, centers=5) 
# n_features=2 (default)
# centers: Số cụm cần tạo
plt.subplot(1, 3, 2)
color = np.array(['r', 'g', 'b', 'y', 'k'])
plt.scatter(X[:, 0], X[:, 1], c=color[y])
print("=================================================")
print(X[:5], y[:5], sep='\n')

# Clustered Dataset Distributed in Circular Fashion 
X, y = make_circles(n_samples=100, noise=0.2) # tạo dữ liệu dạng 2 vòng tròn đồng tâm -> thường dùng cho bài toán phân loại (supervised learning)
plt.subplot(1, 3, 3)
plt.scatter(X[:, 0], X[:, 1], c=color[y])
print("=================================================")
print(X[:5], y[:5], sep='\n')

plt.tight_layout()
```

## **Getting started with `scikit-learn`**

### Read about: **[Data Cleansing](http://localhost:8888/notebooks/Pandas%2Fdataframe%2Fdata_cleansing.ipynb)**

```python
area = [
    [45], [50], [55], [60], [62],
    [65], [70], [75], [78], [80],
    [85], [90], [95], [100], [110],
    [120], [130], [140], [150], [160]
]
# X

house_price = [
    [45000], [52000], [58000], [61000], [64000],
    [70000], [75000], [82000], [85000], [88000],
    [94000], [102000], [109000], [118000], [132000],
    [145000], [158000], [172000], [185000], [198000]
]
#y

plt.scatter(area, house_price, marker='s', c='r')
plt.xlabel('Area(m2)')
plt.ylabel('House Price ($)')
plt.title('House Price based on Area')
```

```python
# Fit the model
from sklearn.linear_model import LinearRegression

# create model
model = LinearRegression()

# fit the data
model.fit(X=area, y=house_price)
# X, y is 2D array
```

```python
# making prediction
# model.predict()
# input is 2d Arr : [[sample_1], [sample2], ...]
# output is 2D arr : [[pred_sample_1], [pred_sample_2], ...]

# single value
print(f'Predict house price of 63m2 area: {round(model.predict([[63]])[0][0], 2)}$') 


# multiple value
areas = [[33], [57], [31], [86], [200]]
pred_prices = model.predict(areas)

for i in range(len(areas)):
    print(f'Area: {areas[i][0]}m2 -----> Predict Price: {round(pred_prices[i][0], 2)}$')

# pred_prices
```

```python
# plot the regression line
plt.scatter(area, house_price, c='r', label='actual house price')
plt.plot(areas, model.predict(areas), 'kD-', label='predict house price')
# plt.axis([10, 100, 10000, 150000]) # x_start, x_end, y_start, y_end
plt.xlabel('Area (m2)')
plt.ylabel('Predict House Price ($)')

plt.legend()
```

```python
# get the slope (hệ số góc, coef, gradient) và intercept (hệ số chặn Y_value khi x (tất cả feature)=0) của regression line
# => y = [slope_]*[x] + intercept_

coef = model.coef_[0][0]
intercept = model.intercept_[0]
print(f'coef: {coef}') # 2D, mỗi phần tử ứng với slope của từng feature
print(f'intercept_: {intercept}') # giá trị y khi tất cả feature = 0

x = 55
print(coef*x + intercept)
print(model.predict([[x]]))
print()
```

```python
# examining the performance of model by calculating the Residual Sum of Squares - RSS / Tổng bình phương sai số giữa dự đoán và thực tế.
# giả sử ta có tập target thật để test: house_price
# tập dữ liệu test: area
# residual : sai số dự đoán
residual = house_price - model.predict(area)
# RSS -> bình phương sai số
RSS = np.sum(residual**2)
print(f"RSS: {RSS}")

# ngoài ra, ta có các chỉ số:
# R^2 // R_square -> Coefficient of Determination : Hệ số xác định
    # R_square = 1 - (RSS/TSS)
        # R_square = 1 -> mô hình dự đoán hoàn hảo, RSS = 0 # lưu ý trường hợp overfitting nếu có nhiều features
        # R_square = 0 -> mô hình không dự đoán được gì (dự đoán = giá trị trùng bình TSS)
        # R_square < 1 -> mô hình dự đoán tệ hơn giá trị trung bình TSS
# TSS - Total Sum of Square : Tổng bình phương biến thiên (Biến thiên của dữ liệu gốc  quanh giá trị trung bình)
TSS = np.sum((house_price - np.mean(house_price))**2)
print(f'TSS: {TSS}')
R_square = 1 - (RSS/TSS)
print(f'R_square: {R_square}')

#scikit-learn cung cap ham tinh r_square:
R_square = model.score(area, house_price)
# model.score(x_test, y_test)
print(f'model.score (R_Square) : {R_square}')
```

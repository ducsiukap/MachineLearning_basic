```python
# display the chart within the same page as front-ends like Jupyter Notebook, .. 
%matplotlib inline

# to plot a chart, using pyplot
import matplotlib.pyplot as plt

# to styling the plot
from matplotlib import style
style.use('ggplot') # 
```

```python
# data
import numpy as np

y = np.random.randint(50, size=10)
X = np.arange(10)
```

# `Plot` - Line Chart

```python
plt.plot(X, y, "go:", label='y')

#  Keyword args:
# - X, y
# - color/c: str 
# - marker: str => kí hiệu điểm #('o', 's', '^', 'v', 'D', 'x', '+', '*')
# - linestyle / ls : str => đường line #('-', '--', ':', '-.', '') # '' for hide the line
#     => format_string: "g--" -> color=g, linestyle=-- / "go--": color=g, marker=o, linestyle=--
# - markersize / ms : float => kích thước marker
# - linewidth / lw : float => độ dày được line
# - alpha: float [0.0, 1.1] => độ trong suốt
# - label: str => tên biểu đồlines


# add another line (plot multiple lines in same chart)
y2 = np.random.randint(100, size=10)
plt.plot(y2, 'rD--', label='y2')

# method:
plt.title("Plot Chart") # chart title
plt.xlabel('X Label') # Ox
plt.ylabel('Y Label') # Oy

plt.legend() # show line label

```

# `Scatter` - Biểu đồ điểm

```python
plt.scatter(X, y, c='g', marker='D', ec='r', lw=1, label='y')

# keyword args:
# - s: kích thước điểm size => (mảng - mỗi điểm 1 kích thước / 1 số cho tất cả)
# - c: màu 
# - marker -> ('o', 's', '^', 'v', 'D', 'x', '+', '*')
# - alpha: độ trong suốt
# - edgecolors / ec: màu viền marker
# - linewidths / lw : độ dày viền
# - cmap (bảng màu nếu c là mảng)

# add another 
plt.scatter(X, y2, c='black', label='y2')

plt.title('Scatter')
plt.xlabel('X')
plt.ylabel('Y')

plt.legend()
```

# `Bar` - Bar Chart / Biểu đồ cột.

# `Pie` - Pie Chart / Biểu đồ tròn

```python
labels=['Chrome', 'Safari', 'Edge', 'Firefox', 'Samsung Internet', 'Opera']
marketshare= [69.23, 14.98, 5.03, 2.26, 1.97, 1.85]

pie = plt.pie(marketshare, labels=labels, autopct='%.1f%%', shadow=True, explode=(0.2, 0.1, 0, 0, 0.3, 0))

# X: dữ liệu
# labels: nhãn cho từng dữ liệu
# colors: danh sách màu cho từng labels
# autopct: str -> formart hiển thị %: '%1.1f%%' -> 1 chữ số tập phân, '%d%%' -> số nguyên
# startangle: góc bắt đầu  (để quay biểu đồ)
# shadow: boolean -> đổ bóng ?
# explode: list -> mieensg nào tách ra => làm nổi bật phần nào đó (0-1.0)
# counterClock: True/false -> vẽ ngược chiều kim đồng hồ.

# turns off the axis lines and labels
plt.axis('equal') #equal, scaled, tight, auto, off
# equal: X=Y => hình ra tròn.

# .pie() return (patches, texts, autotexts)
# - patches: list of matplotlib.patches.Wedge. => Dùng để tùy chỉnh màu, viền, alpha,... từng miếng.
# - texts : matplotlib.text.Text object for labels
# - autotexts: Text object for autopct or None
print(">>> Patches: ", pie[0])

plt.title('Browser marketshare')

# display the legend
# dùng khi không muốn dùng nhãn trực tiếp trên từng miếng .pie(labels=None)
plt.legend(pie[0], labels, loc=0)
# patches = pie[0] => danh sách Wedge object 
# labels: Nhãn tương ứng
#           => note: patches và labels phải cùng xuất hiện!.
# loc='best' / loc=0 => legend không che pie
        # ’best’ 0
        # ’upper right’ 1
        # ’upper left’ 2
        # ’lower left’ 3
        # ’lower right’ 4
        # ’right’ 5
        # ’center left’ 6
        # ’center right’ 7
        # ’lower center’ 8
        # ’upper center’ 9
        # ’center’ 10
```

# `hist()` - Histogram chart / Biểu đồ tần xuất

```python
X = np.random.randint(20, size=100)
for num in set(X):
    print(f'{num} - count: {np.sum(X==num)}')
    
plt.hist(X, bins=20, color='g', edgecolor='black', density=False)
# X: dữ liệu số
# bins: Số cột
# color: màu cột
# alpha
# label
# density=True -> chuẩn thóa thành tỷ lệ %
# cumulative: False / True -> tổng tích lũy từ nhóm trước
```

```python
# histogram nhiều nhóm
X2 = np.random.randint(20, size=100)

# List X nhận vào nhiều mảng
plt.hist([X, X2], color=['g', 'r'], label=['X', 'X2'])
plt.legend()
```

# `Boxplot` - Boxplot / Biểu đồ hộp

```python
data1 = np.random.normal(100, 10, 200)  # nhóm 1
data2 = np.random.normal(80, 20, 200)   # nhóm 2
data3 = np.random.normal(90, 15, 200)   # nhóm 3

data = [data1, data2, data3]
labels = ['Nhóm 1', 'Nhóm 2', 'Nhóm 3']

plt.boxplot(data, tick_labels=labels, patch_artist=True, showmeans=True)
# X : dữ liệu
# notch: True/False  -> rãnh giữ hộp
# vert: True (dọc) / False (ngang)
# patch_artist=True -> tô màu hộp
# tick_labels -> nhãn từng box
# showmeans=True -> hiện giá trị trung bình
# meanline => vẽ line cho mean (thay vì maker)
```

# `Subplot` - plot multiple charts in a figure

```python
y = np.random.randint(10, size=5).reshape(-1)
X = [1, 2, 3, 4, 5]

# .subplot(nrows, ncols, index)
plt.subplot(1, 3, 1)
plt.plot(X, y)
plt.title('Line Chart')

plt.subplot(1, 3, 2)
plt.scatter(X, y)
plt.title('Scatter Chart')

plt.subplot(1, 3, 3)
plt.bar(X, y)
plt.title('Bar Chart')

plt.tight_layout() # tự động căn chỉnh để không bị đè label/title
```

# `Subplots()` - modern way to display multiple chart

```python
y = np.random.randint(10, size=5).reshape(-1)
X = [1, 2, 3, 4, 5]

# .subplots() -> return : figures & axes array
fig, ax = plt.subplots(2, 2) # tạo lưới 2x2

# to draw
ax[0, 0].plot(X, y)
# customize the chart
ax[0, 0].set_title('Line Chart')
# ax[0, 0].set_xlabel()
# ax[0, 0].set_ylabel()

ax[0, 1].bar(X, y)
ax[1, 0].scatter(X, y)
ax[1, 1].pie(y, labels=X)

# tight_layout()
fig.tight_layout()
```

# `figure` - another way to draw multiple diagram

```python
fig = plt.figure(figsize=(12, 4))
# figure

# add subplot
ax = fig.add_subplot(1, 3, 1)
ax.plot(X, y)

ax = fig.add_subplot(1, 3, 2)
ax.scatter(X, y)

ax = fig.add_subplot(1, 3, 3)
ax.bar(X, y)
```

# 3D Plotting

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure(figsize=(20, 10))

# create 3d axes
# plt.axes(projection='3d')
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot(x, y, z)

ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.scatter(x, y, z)

plt.tight_layout()
```

- `ax=plt.gca()` -> lấy axes hiện tại đang được vẽ - get current axes
